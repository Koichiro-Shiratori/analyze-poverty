```python
import pandas as pd

df1 = pd.read_csv('../../data/www.e-stat.go.jp/c01.csv', encoding='shift-jis')
# csvから最新のデータのみを抽出する
# indexをjapanmapに与えるためにshiftでズラす
df1 = df1.shift(-933)[0:47]
df1
```

```python
# 人口（総数）は文字列型なのでintに変換する
x = df1['人口（総数）'].astype(int)
x
```

```python
# 割合に変換する
x = (x - min(x)) / (max(x) - min(x))
print(x)

```

```python
import matplotlib.pyplot as plt

cmap = plt.get_cmap('Reds')

def tint(t):
    return tuple(int(_) for _ in t)

# RGBのtupleに変換する
d = {i: tint(cmap(x[i], bytes=True)[:3]) for i in range(0,46)}
d
```

```python
from japanmap import picture

plt.imshow(picture(d))
plt.tick_params(bottom=False, top=False, left=False, right=False,
                labelbottom=False, labeltop=False, labelleft=False, labelright=False)
```
