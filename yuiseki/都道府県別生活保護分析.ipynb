```python
import pandas as pd
import numpy as np

df1 = pd.read_excel('../data/www.mhlw.go.jp/2020/都道府県別被保護者調査.xls')
df1 = df1.shift(-5)[0:47]
columns = ['無１', '都道府県', '無２', '被保護世帯総数', '現に保護を受けた世帯', '保護停止中の世帯', '被保護実人員総数', '現に保護を受けた人数', '保護停止中の人数', '保護開始世帯数', '保護廃止世帯数']
df1.columns = columns
df1
```

```python
x1 = df1['被保護世帯総数']
x1 = (x1 - min(x1)) / (max(x1) - min(x1))
x1
```

```python
import matplotlib.pyplot as plt
from japanmap import picture

cmap = plt.get_cmap('Reds')

def tint(t):
    return tuple(int(_) for _ in t)

d = {i: tint(cmap(x1[i], bytes=True)[:3]) for i in range(0,47)}

plt.figure(figsize = (10,10))
plt.imshow(picture(d))
plt.tick_params(bottom=False, top=False, left=False, right=False,
                labelbottom=False, labeltop=False, labelleft=False, labelright=False)
```

```python
df2 = pd.read_csv('../data/www.e-stat.go.jp/c01.csv', encoding='shift-jis')
df2 = df2.shift(-933)[0:47]
df2
```

```python
df3 = pd.merge(df1, df2, how='left', left_index=True, right_index=True)
df3
```

```python
df4 = pd.DataFrame(df3, columns=['都道府県名', '被保護実人員総数', '人口（総数）'])
df4['生活保護比率'] = df4['被保護実人員総数']/df4['人口（総数）'].astype(int)
df4
```

```python
x2 = df4['生活保護比率']
x2 = (x2 - min(x2)) / (max(x2) - min(x2))
x2
```

```python
cmap = plt.get_cmap('Blues')

def tint(t):
    return tuple(int(_) for _ in t)

d = {i+1: tint(cmap(x2[i], bytes=True)[:3]) for i in range(0,46)}

plt.figure(figsize = (10,10))
plt.imshow(picture(d))
plt.tick_params(bottom=False, top=False, left=False, right=False,
                labelbottom=False, labeltop=False, labelleft=False, labelright=False)
```
