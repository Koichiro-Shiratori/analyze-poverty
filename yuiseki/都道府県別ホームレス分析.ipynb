```python
import pandas as pd

df1 = pd.read_csv('../data/www.mhlw.go.jp/2019/都道府県別ホームレス数.csv')
df1
```

```python
x1 = df1['合計人数']
x1 = (x1 - min(x1)) / (max(x1) - min(x1))
x1
```

```python
import matplotlib.pyplot as plt
from japanmap import picture
cmap = plt.get_cmap('Reds')

def tint(t):
    return tuple(int(_) for _ in t)

d = {i+1: tint(cmap(x1[i], bytes=True)[:3]) for i in range(0,46)}

plt.figure(figsize = (10,10))
plt.imshow(picture(d))
plt.tick_params(bottom=False, top=False, left=False, right=False,
                labelbottom=False, labeltop=False, labelleft=False, labelright=False)
```

```python
df2 = pd.read_csv('../data/www.e-stat.go.jp/c01.csv', encoding='shift-jis')
df2 = df2.shift(-933)[0:47]
df2
```

```python
df3 = pd.merge(df1, df2, on='都道府県名')
df3
```

```python
df4 = pd.DataFrame(df3, columns=['都道府県名', '合計人数', '人口（総数）'])
df4['ホームレス比率'] = df4['合計人数']/df4['人口（総数）'].astype(int)
df4
```

```python
x2 = df4['ホームレス比率']
x2 = (x2 - min(x2)) / (max(x2) - min(x2))
x2
```

```python
cmap = plt.get_cmap('Blues')

def tint(t):
    return tuple(int(_) for _ in t)

d = {i+1: tint(cmap(x2[i], bytes=True)[:3]) for i in range(0,46)}

plt.figure(figsize = (10,10))
plt.imshow(picture(d))
plt.tick_params(bottom=False, top=False, left=False, right=False,
                labelbottom=False, labeltop=False, labelleft=False, labelright=False)
```
