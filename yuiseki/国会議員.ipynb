```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import japanize_matplotlib
ttf_path = japanize_matplotlib.get_font_ttf_path()

データURL = "https://smartnews-smri.github.io/house-of-councillors/data/giin.csv"
df = pd.read_csv(データURL, encoding='utf-8')

fig, ax = plt.subplots(facecolor='lightslategray')

df['会派'].value_counts().plot(
  kind="pie",
  figsize=(10, 10),
  autopct='%.f%%',
  startangle=90,
  counterclock=False,
  ax=ax
)
ax.set_facecolor("lightslategray")
```
